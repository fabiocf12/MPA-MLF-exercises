```python
#1.3
def numb_to_bin(num : int) -> None:

  result = ""

  while num > 0:
    result = str(num % 2) + result
    num = num // 2

  print(result)

numb_to_bin(12784)
```

```python
#1.5
import random

def rock_paper_scissors(n_rounds : int) -> None:

    rounds = n_rounds
    user_score = computer_score = 0

    while n_rounds > 0:

        user_choice = str(input("\nEnter a choice (rock, paper, scissors): ")).lower()
        computer_choice = random.choice(["rock", "paper", "scissors"])

        print(f"\nMy choice was {computer_choice}!")

        if user_choice == computer_choice:
            print("We tie this round!")

        elif user_choice == "rock":
            if computer_choice == "scissors":
                print("You win this round!")
                user_score += 1
            else:
                print("You lose this round!")
                computer_score += 1

        elif user_choice == "paper":
            if computer_choice == "rock":
                print("You win this round!")
                user_score +=1
            else:
                print("You lose this round!")
                computer_score += 1

        elif user_choice == "scissors":
            if computer_choice == "paper":
                print("You win this round!")
                user_score += 1
            else:
                print("You lose this round!")
                computer_score += 1
        else:
          print("Your input is not valid!!!\n")

        n_rounds -=1
        print(f"** ROUND {rounds - n_rounds} **\n**Computer Score {computer_score} **\n**User Score {user_score} **\n")

    if computer_score > user_score:
        print("You lose")
    elif computer_score < user_score:
        print("You win")
    else:
      print("It is a tie")

rock_paper_scissors(5)
```

```python
#2.1
import numpy as np
import time
def create_array_nxn(n : int) -> np.array:

  if n <= 0:
    print("Invalid input")
    return

  matrix = np.arange(n ** 2 - 1,-1,-1)
  matrix = matrix.reshape(n,n)
  print(matrix)
  return matrix

def apply_threshold_loop(matrix : np.array, threshold: int) -> np.ndarray:

  for i in range(len(matrix)):
    for k in range(len(matrix[i])):
      if matrix[i][k] < threshold:
        matrix[i][k] = 0
  print(f"\n RESULTS WITH LOOP\n{matrix}")

  return matrix

def apply_threshold_vectorized(matrix : np.array, threshold: int) -> np.ndarray:

  matrix[matrix < threshold] = 0
  print(f"\n RESULTS WITH BUILT-IN FUNC\n{matrix}")

  return matrix

def compare_performance(n: int, threshold: int) -> None:

  matrix = create_array_nxn(1000)

  start = time.time()
  apply_threshold_loop(matrix.copy(),threshold)
  end = time.time()
  print(f"\nTime with loop (seconds): {end - start}")

  start1 = time.time()
  apply_threshold_vectorized(matrix.copy(),threshold)
  end1 = time.time()
  print(f"\nTime with built-in func (seconds): {end1 - start1}")


compare_performance(1000,30)
```
